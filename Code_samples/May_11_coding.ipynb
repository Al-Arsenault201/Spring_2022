```python
mydir = "/Users/alfredarsenault/Documents/GitHub/Spring_2022/Project_3/Data_files"
import os  #import the python operating system library
concatenated_data_list = []
for filename in os.listdir(mydir):
    print(filename)
    if "review" in filename:
        real_dir = mydir + "/" + filename   #gives the full pathname of the file
        with open ( real_dir, "r") as infile:
            data = infile.read()
            concatenated_data_list.append(data)

#print(len(concatenated_data_list))
#for i in range(len(concatenated_data_list)):
#    print(concatenated_data_list)
        
    
```

```python
term_list = [ []]

data_list = concatenated_data_list[0].split()
#print(data_list)
    
term_list =[ [data_list[0],1]]
term_list.append([data_list[1],1])

for i in range(2, len(data_list)):
    found = False
    for j in range(len(term_list)):
        if data_list[i] == term_list[j][0]:
            found = True
            term_list[j][1] += 1
    if not found:
        term_list.append([data_list[i],1])

for i in range(len(term_list)):
    term_list[i][1] /= len(data_list)
    
print(term_list)
```

```python
stopwords = ["stop","the","to","and","a","in","it","is","i","that","bad","on","for","were","was"]
print("initial term list length", len(term_list))
edited_term_list = []
for i in range(len(term_list)):
    if not (term_list[i][0].lower() in stopwords):
        edited_term_list.append(term_list[i])
print("final term list length", len(edited_term_list))
```
