Now that we have read in a review, we need to create a Term Frequency (TF) table for this file

```python
data_list = data.split()
for i in range(len(data_list)):
    print(data_list[i])
```

create a 2D list for TF

```python
term_list = [ [data_list[0],1]  ]
for i in range(1,len(data_list)):
        found = False
        for j in range(len(term_list)):
             if data_list[i] == term_list[j][0]:
                 term_list[j][1] +=1
                 found = True
        if not found:
             term_list.append([data_list[i],1])
             

for k in range(len(term_list)):
    print(term_list[k])
             
```

## Stopwords
for the purposes of this demonstration, we're going to create a list of stopwords.<br>
In the assignment, you'll read in the file and create the list by splitting


```python
stopwords = ["a", "an", "and", "the"]
for i in stopwords:
    #check to see if the word is in the term_list
    for j in range(len(term_list)):
        if i == term_list[j][0]:
            term_list = term_list.pop(j)
            
            
print(len(term_list))
```

```python
import pandas
from plotnine import ggplot, aes, geom_point, geom_bar, geom_col
```

```python
df = pandas.DataFrame(term_list)
df.columns=("Word","Frequency")
```

```python
print(df.head(20))
```

Now, our data is in a dataframe; it's time to create a plot. <br>
