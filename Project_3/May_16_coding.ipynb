We printed out the term_list table.<br><br> Now go ahead and remove the stopwords from the entire table

```python
stopwords = ["stop","the","to","and","a","in","it","is","i","that","bad","on","for","were","was"]
print("initial term list length", len(edited_term_list))
final_term_list = []
for i in range(len(edited_term_list)):
    if not (edited_term_list[i][0].lower() in stopwords):
        final_term_list.append(edited_term_list[i])
print("final term list length", len(final_term_list))
```

Now calculate the IDF values and build the list. Remember that the variable we're working with now is "final_term_list." That's our actual TF list.

```python
import math
idf = []
for i in range(len(final_term_list)):
    num_occurrences = 0
    for j in range(1,len(final_term_list[i])):
        if final_term_list[i][j] > 0:
            num_occurrences += 1
#    print(num_occurrences)
    inverse_doc_freq = math.log(10/num_occurrences)
    new_item = [final_term_list[i][0], inverse_doc_freq]
    idf.append(new_item)
    
for j in range(len(idf)):
    print(idf[j])
    
```

So there's the IDF table. <br> 
Now, to calculate the TF-IDF, just multiply each number in the TF table (final_term_list) by the IDF for that same word. 

```python
tf_idf = []
for i in range(len(final_term_list)):
    new_row = [final_term_list[i][0]]
    for j in range(1,len(final_term_list[i])):
        new_row.append(final_term_list[i][j] * idf[i][1])
    tf_idf.append(new_row)

for k in range(len(tf_idf)):
    print(tf_idf[k])
    
```
